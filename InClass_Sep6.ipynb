```python
import pandas as pd
import numpy as np
import dmba 
import seaborn as ss
import matplotlib.pyplot as plt
import calendar
```

```python
housing_df = dmba.load_data('BostonHousing.csv')
housing_df = housing_df.rename(columns = {'CAT. MEDV': 'CAT_MEDV'})

housing_df.head(9)
```

```python
Amtrak_df = dmba.load_data('Amtrak.csv')
Amtrak_df['Date'] = pd.to_datetime(Amtrak_df.Month, format='%d/%m/%Y')
ridership_ts = pd.Series(Amtrak_df.Ridership.values, index = Amtrak_df.Date)
ridership_ts
```

```python
#pandas version of plotting
ridership_ts.plot(ylim = [1300, 2300], legend=False, figsize=[5,3])
plt.xlabel('Year')
plt.ylabel('Ridership (in 000s)')

plt.tight_layout()
plt.show()
```

```python
housing_df.plot.scatter(x='LSTAT', y='MEDV', legend=False)
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(5, 3)

ax.scatter(
    housing_df.LSTAT,
    housing_df.MEDV,
    color='C2',
    facecolor='none'
)
```

```python
ax=housing_df.groupby('CHAS').mean().MEDV.plot(
    kind='bar',
    figsize=[5,3]
)

ax.set_ylabel('Avg. MEDV')
plt.tight_layout()
```

```python
dataForPlot = housing_df.groupby('CHAS').mean()['CAT_MEDV'] *100
ax = dataForPlot.plot(kind='bar', figsize=[5,3]) #ax.bar(dataForPlot.index, dataForPlot, color=['C5', 'C1']) does something similar
ax.set_ylabel('% of CAT.MEDV')
```

```python
ax = housing_df.MEDV.hist()
ax.set_xlabel('MEDV')
ax.set_ylabel('count')
```

```python
ax = housing_df.boxplot(
    column='MEDV',
    by='CHAS'
)

ax.set_ylabel('MEDV')
plt.suptitle('aaa')
plt.title('bbb')
```
